```python
'''Handling missing values: 

 1)  fillna() method 
 2)  Forward fill method - fillna(method="ffill",limit=m)
 3)  Backward fill method - df2=df.fillna(method="bfill")
 4)  dropna() method
 '''
df2 = df1.fillna(method ='ffill')
```

```python
df2.isnull().sum()
```

```python
#Droping all the rows which contains all the values NaN
df3 = df2.dropna(how ='all')
```
