```python
Cities =[]
Years =[]
for game in open(r"C:\Users\v-trugan\Downloads\Ex_Files_Programming_Efficiently\games.txt",'r'):
    words = game.split()
    city = ''.join(words[:-1])
    years = words[-1].strip('()')
    print('City:' , city, "Years:", years)
    Cities.append(city)
    Years.append(years)
```

```python
print(Cities,Years )
```

```python
for year, city in zip(Years, Cities):
    print(year, city)
```

```python
geolocator = geopy.geocoders.Nominatim(user_agent = 'programming-efficiently')
```

```python
locations = {}
for city in Cities:
    print('Locating: ', city)
    
    locations[city] = geolocator.geocode(city.split('/')[0])
```

```python
import json
```

```python
coordinates_by_city =json.load(open(r"C:\Users\v-trugan\Downloads\Ex_Files_Programming_Efficiently\coords.json", 'r'))
coordinates_by_city
```

```python
plt.figure(figsize = (10,5))
axes = plt.axes(projection = cartopy.crs.PlateCarree())
axes.add_feature(cartopy.feature.COASTLINE, linewidth = 0.5)
axes.add_feature(cartopy.feature.BORDERS, linewidth = 0.5)

for L in locations.values():
    axes.plot(L.longitude, L.latitude, 'r.')
    
plt.axis(xmin = -180, xmax= 180, ymin = -90, ymax = 90)    
    
```

```python
cities_by_years = {city: year for city, year in zip(Years, Cities) }
```

```python
cities_by_years
```

```python
result = {city for year, city in cities_by_years.items() if int(year) > 1930}
result 
```

```python
# start a large figure
plt.figure(figsize=(10,10))

# make the projection
axes = plt.axes(projection=cartopy.crs.Orthographic(central_latitude=45, central_longitude=0))

# add boundaries
axes.add_feature(cartopy.feature.COASTLINE, linewidth=0.5)
axes.add_feature(cartopy.feature.BORDERS, linewidth=0.5)

for year, city in cities_by_years.items():        
    lon, lat = coordinates_by_city[city]

    # skip cities on the other side of the globe
    if lon < 130:
        # plot with cartopy coordinate system
        axes.plot(lon, lat, 'r.', transform=cartopy.crs.Geodetic())
        axes.text(lon, lat, year, transform=cartopy.crs.Geodetic())
    
axes.set_global()
```
