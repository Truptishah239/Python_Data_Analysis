```python
def findmedal(year =None, event = None, team = None):
    filtered = medals
    
    if year is not None:
        filtered = [medal for medal in filtered if medal.year == year]
        
    if event is not None:
        filtered = [medal for medal in filtered if medal.event == event]
        
    if team is not None:
        filtered = [medal for medal in filtered if medal.team == team]
        
    return filtered    
```

```python
findmedal(year = '1896', team = 'USA')
```

```python
def findmedal2(**kwargs):
    filtered = medals
    for key, value in kwargs.items():
        filtered = [medal for medal in filtered if getattr(medal, key) == value]
      
    return filtered
```

```python
findmedal2(year='1896',team='USA')
```

```python
def normalize(events):
    return ''.join(word for word in events.split() if word not in ['men', 'women']) 
```

```python
events_by_athlete = collections.defaultdict(set)

for medal in medals:
    events_by_athlete[medal.athlete].add(normalize(medal.event))
    
{athlete: events for athlete, events in events_by_athlete.items() if len(events) >= 4}
```
